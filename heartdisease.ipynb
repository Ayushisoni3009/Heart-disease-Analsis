Table Discription:

1. age
2. sex

3. chest pain type (4 values)

3a. Value 0: typical angina

3b. Value 1: atypical angina

3c. Value 2: non-anginal pain

3d. Value 3: asymptomatic

4. trestbps: resting blood pressure (in mm Hg on admission to the hospital)

5. chol: serum cholestoral in mg/dl

6. fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)

7. restecg: resting electrocardiographic results
 7a. Value 0: normal

7b.  Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevatic depression of > 0.05 mV)

7c. Value 2: showing probable or definite left ventricular hypertrophy by Estes' crit

8. thalach: maximum heart rate achieved

9.  exang: exercise induced angina (1 = yes; 0 = no)

10.  oldpeak = ST depression induced by exercise relative to rest

11.  slope: the slope of the peak exercise ST segment

11a.  Value 1: upsloping

11b.  Value 2: flat

11c.  Value 3: downsloping

12.  ca: number of major vessels (0-3) colored by flourosopy

13. thal: 3 = normal; 6 = fixed defect; 7 = reversable defect

14.  target: 0=less chance of heart attack, 1=more chance of heart attack

# 19. comparasion of resting blood pressure as per sex

```python
import pandas as pd
data = pd.DataFrame({
    'trestbps': [120, 130, 140, 150, 160, 170, 180, 190],
    'sex': [0, 1, 0, 1, 0, 1, 0, 1]
})


g = sns.FacetGrid(data, hue="sex", height=6, aspect=4)
g.map(sns.kdeplot, 'trestbps', fill=True)  # Use fill=True instead of shade=True


plt.xlabel('Resting Blood Pressure (trestbps)')
plt.ylabel('Density')
plt.title('Kernel Density Estimate of Resting Blood Pressure by Sex')
plt.show()
```
